```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

def main():
    while True:
        x = int(input())
        if x > 0:
            break
    print("YES" if is_prime(x) else "NO")

if __name__ == "__main__":
    main()
```

```python
def sieve(n):
    if n < 2:
        return []
    p = [True] * (n + 1)
    p[0] = p[1] = False
    i = 2
    while i * i <= n:
        if p[i]:
            for j in range(i*i, n+1, i):
                p[j] = False
        i += 1
    return p

def main():
    import sys
    data = sys.stdin.read().strip().split()
    nums = list(map(int, data))
    pairs = [(nums[i], nums[i+1]) for i in range(0, len(nums), 2)]

    maxb = max(b for _, b in pairs)
    p = sieve(maxb)

    prefix = [0]*(maxb+1)
    s = 0
    for i in range(maxb+1):
        if p[i]:
            s += i
        prefix[i] = s

    for a,b in pairs:
        ans = prefix[b] - (prefix[a-1] if a > 0 else 0)
        print(ans)

if __name__ == "__main__":
    main()
```

```python
def main():
    n = int(input())
    x = n
    largest = -1

    while x % 2 == 0:
        largest = 2
        x //= 2

    f = 3
    while f * f <= x:
        while x % f == 0:
            largest = f
            x //= f
        f += 2

    if x > 1:
        largest = x
    print(largest)

if __name__ == "__main__":
    main()
```

```python
def reverse_int(n):
    return int(str(n)[::-1])

def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

def main():
    n = int(input())
    steps = 0
    while not is_palindrome(n):
        n = n + reverse_int(n)
        steps += 1
    print(steps, n)

if __name__ == "__main__":
    main()
```

```python
def distinct_digits(n):
    s = str(n)
    return len(s) == len(set(s))

def main():
    n = int(input())
    res = []
    i = 1
    while i * i <= n:
        sq = i * i
        if distinct_digits(sq):
            res.append(str(sq))
        i += 1
    
    if res:
        print(" ".join(res))
    else:
        print("(no number)")

if __name__ == "__main__":
    main()
```

```python
def collatz_length(x, memo):
    if x in memo:
        return memo[x]
    if x == 1:
        memo[x] = 1
        return 1
    if x % 2 == 0:
        memo[x] = 1 + collatz_length(x//2, memo)
    else:
        memo[x] = 1 + collatz_length(3*x + 1, memo)
    return memo[x]

def main():
    n = int(input())
    memo = {}
    best_x = 1
    best_len = 1
    for x in range(1, n+1):
        l = collatz_length(x, memo)
        if l > best_len or (l == best_len and x < best_x):
            best_len = l
            best_x = x
    print(best_x, best_len)

if __name__ == "__main__":
    main()
```

```python
def main():
    n = int(input())
    if n % 2 != 0:
        print(0)
        return
    m = n // 2
    cnt = 0
    i = 1
    while i * i <= m:
        if m % i == 0:
            cnt += 1
            if i != m // i:
                cnt += 1
        i += 1
    print(cnt)

if __name__ == "__main__":
    main()
```

```python
def sum_div_except_self(n):
    if n <= 1: return 0
    s = 1
    i = 2
    while i * i <= n:
        if n % i == 0:
            s += i
            if i != n//i:
                s += n//i
        i += 1
    return s

def main():
    a, b = map(int, input().split())
    print("true" if sum_div_except_self(a) == b and sum_div_except_self(b) == a else "false")

if __name__ == "__main__":
    main()
```

```python
def main():
    T, L = map(int, input().split())
    if (L - 2*T) % 2 != 0:
        print("invalid")
        return
    D = (L - 2*T)//2
    G = T - D
    if G < 0 or D < 0:
        print("invalid")
    else:
        print(G, D)

if __name__ == "__main__":
    main()
```

```python
#W3A11
a = int(input("Nhập cạnh a: "))
b = int(input("Nhập cạnh b: "))
c = int(input("Nhập cạnh c: "))

if a + b <= c or a + c <= b or b + c <= a:
    print("Không phải tam giác")
else:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
```

```python
#W3A17

a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
c = int(input("Nhập c: "))
d = int(input("Nhập d: "))

if a == 0:
    if b == 0 and c == 0 and d == 0:
        print("Là cấp số nhân (r = 0)")
    else:
        print("Không phải cấp số nhân")
else:
    if b % a == 0:     
        r = b // a     
        if b * r == c and c * r == d:
            print("Là cấp số nhân, công bội =", r)
        else:
            print("Không phải cấp số nhân")
    else:
        print("Không phải cấp số nhân")
```

```python
def Max_Of_Two(a, b) -> int:
    return a if a > b else b
```

```python
def swap(a, b):
    return b, a
```

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
def is_perfect(n: int) -> bool:
    if n <= 1:
        return False
    s = 1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            s += i
            if i * i != n:
                s += n // i
    return s == n
```

```python
def find_position(lst, k: int) -> int:
    for i, v in enumerate(lst):
        if v == k:
            return i
    return -1
```

```python
def factorial(n: int) -> int:
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        result = num1 / num2
    else:
        raise ValueError("Toán tử không hợp lệ!")
    return round(result, 2)
```

```python
def hamming_distance(x: int, y: int) -> int:
    return bin(x ^ y).count("1")
```

```python
def sum_of_digits(n: int) -> int:
    return sum(int(d) for d in str(n))
```

```python
def is_isomorphic(a: str, b: str) -> bool:
    map_a = {}
    map_b = {}
    for x, y in zip(a, b):
        if (x in map_a and map_a[x] != y) or (y in map_b and map_b[y] != x):
            return False
        map_a[x] = y
        map_b[y] = x
    return True
```
